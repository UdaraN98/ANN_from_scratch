```python
#creating a data loading function

import os
import numpy as np

def load_data(bp_path, sz_path):
    bp_data1 = []
    bp_data2 = []
    for subdir, _, files in os.walk(bp_path):
        for file in files:
            file_path = os.path.join(subdir, file)
            if file_path.endswith('.npy'):
                data = np.load(file_path)
                bp_data1.append(data[:, :5640])  # First input from BP folder
                bp_data2.append(data[:, 5640:])  # Second input from BP folder
    bp_data1 = np.array(bp_data1)
    bp_data2 = np.array(bp_data2)

    
    sz_data1 = []
    sz_data2 = []
    for subdir, _, files in os.walk(sz_path):
        for file in files:
            file_path = os.path.join(subdir, file)
            if file_path.endswith('.npy'):
                data = np.load(file_path)
                sz_data1.append(data[:, :5640])  # First input from BP folder
                sz_data2.append(data[:, 5640:])  # Second input from BP folder
    sz_data1 = np.array(sz_data1)
    sz_data2 = np.array(sz_data2)

    # Create labels
    bp_labels = np.ones(bp_data1.shape[0], dtype=int)
    sz_labels = np.zeros(sz_data1.shape[0], dtype=int)

    # Combine data and labels
    data1 = np.concatenate([bp_data1, sz_data1], axis=0)
    data2 = np.concatenate([bp_data2, sz_data2], axis=0)
    labels = np.concatenate([bp_labels, sz_labels], axis=0)

    return data1, data2, labels
```

```python
in1,in2, labels = load_data("/content/kaggle_data/data/train/BP","/content/kaggle_data/data/train/SZ")
```

```python
#creating a simple nn

import tensorflow as tf
import numpy as np

# Load the input data
input1 = np.load('input1.npy')  # shape: (5640, 105)
input2 = np.load('input2.npy')  # shape: (5640, size)

# Define the neural network architecture
input1_layer = tf.keras.layers.Input(shape=(105,))
input2_layer = tf.keras.layers.Input(shape=(1,))
hidden_layer = tf.keras.layers.Concatenate()([input1_layer, input2_layer])
hidden_layer = tf.keras.layers.Dense(64, activation='relu')(hidden_layer)
output_layer = tf.keras.layers.Dense(1, activation='sigmoid')(hidden_layer)

# Define the model
model = tf.keras.Model(inputs=[input1_layer, input2_layer], outputs=output_layer)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x=[input1, input2], y=labels, epochs=10, batch_size=32)
```
